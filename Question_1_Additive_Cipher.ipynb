Question 1 : Write a program that can encrypt  and decrypt using the Additive Cipher.

```python
#Python program to illustrate Additive Cipher Technique
#In this technique each letter of a given text is replaced by a letter some fixed number of positions down the alphabet.
#Eg. With a shift of 2, A would be replaced by C, B would become D, and so on.

#Defining a function named encrpt that would take text to be encrypted and shift integer(number of positions a letter is to be sifted) as input 

def encrypt(msg,s):
    """
    Input:
      msg: This is the msg to be encrypted.
      s: This is the number of positions a letter is to be shifted to achieve encryption
    Output:
      The encrypted text corresponding to the inputs given.
    Eg:
      Msg  : ATTACK
      S : 1
      Encrypted text: BUUBDL
    """
    result = "" 
 
    # traverse msg letter by letter
    for i in range(len(msg)):
        char = msg[i]
 
        # Encrypt uppercase characters
        if (char.isupper()):
            result += chr((ord(char) + s-65) % 26 + 65)
 
        # Encrypt lowercase characters
        else:
            result += chr((ord(char) + s - 97) % 26 + 97)
 
    return result
 
#Defining a function named decrpt that would take text to be decrypted and shift integer(number of positions a letter was sifted down by encrypt) as input 
def decrypt(msg,s):
    """
    Input:
      msg: This is the msg to be decrypted.
      s: This is the number of positions a letter is shifted while encryption
    Output:
      The decrypted text corresponding to the inputs given.
    Eg:
      Msg  : BUUBDL
      S : 1
      Encrypted text: ATTACK
    """
    s=26-s  #the cyclic property of the cipher under modulo 26 
    result = "" 
 
    # traverse msg letter by letter
    for i in range(len(msg)):
        char = msg[i]
 
        # Encrypt uppercase characters
        if (char.isupper()):
            result += chr((ord(char) + s-65) % 26 + 65)
 
        # Encrypt lowercase characters
        else:
            result += chr((ord(char) + s - 97) % 26 + 97)
 
    return result

#defining main function    
def main():
  print("\nAdditive Cipher...")
  text = "HellomynameisAliceIliveinherefromquiteawhile"
  s = 2
  print ("\nMessage  : " + text)
  print ("\nShift : " + str(s))
  print ("\nEncrypted message : " + encrypt(text,s))
  print ("\nDecrypted message : " + decrypt(encrypt(text,s),s))

if __name__=="__main__":
    main()
```
